```python
main_fan_type = []
for fan in range(5,89):
    main_fan_type+=fan_point_dict[fan]
```

```python
dst = "../processed_fan_morph"
dir_list = os.listdir(dst)
pbar = tqdm(dir_list)
morphing = defaultdict(default_value)
for file in pbar:
    (
        init_dist,  # 1
        init_final_target_diff,  # 2
        tile_in_init_target,  # 3
        tile_in_final_target,  # 4
        tile_diff_between_targets,  # 5
        init_fan_type_and_count,  # 6
        final_fan_type_and_count,  # 7
        winner_id,  # 8
    ) = postprocess_fan_morph.load_from_result(dst, file)
    for id in range(4):
        for init_fan in init_fan_type_and_count[id]:
            if init_fan in main_fan_type:
                for final_fan in final_fan_type_and_count[id]:
                    if final_fan in main_fan_type:
                        key = init_fan+'-'+final_fan
                        morphing[key]+=1
```

```python
morphing_sorted = dict(sorted(morphing.items(), key=lambda item: -item[1]))
```

```python
morphing_cleaned = defaultdict(default_list)
for key in morphing_sorted:
    fans = key.split("-")
    init_fan = fans[0]
    final_fan = fans[1]
    if init_fan!=final_fan and morphing_sorted[key]>1000:
        morphing_cleaned[init_fan].append('{}({})'.format(final_fan, morphing_sorted[key]))
```
